```python
import urllib.request 
from selenium import webdriver
import requests
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
import csv
from bs4 import BeautifulSoup

f = csv.writer(open('Campinas.csv', 'w'))
f.writerow(['Nome', 'Numero', 'Valor'])

html = 'https://www.tripadvisor.com.br/Search?q=Doce&searchSessionId=DE0ED12DFF158D4E75558992113D58AB1624644553153ssid&sid=E282C66D2D5C4440ACA182320D6F0C1D1624644696901&blockRedirect=true&ssrc=m&geo=303605&rf=2'

def linksDosEstabelecimentos(htmlV):
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.implicitly_wait(0.5)
    driver.get(htmlV)

    valor = driver.find_element_by_id('_evidon-accept-button')
    valor.click()

    print('Click foi')

    #driver.close()

    vetVal= []

    for i in range(1,2):

        if i == 1:
            clas = driver.find_element_by_xpath('//*')

            valueCalss = clas.get_attribute('outerHTML')

            value = BeautifulSoup(valueCalss, 'html.parser')

            links = value.find_all(class_='review_count')

            for link in links:
                #for value in link:
                #value = link.find('a')
                val = link.get('href')

                print(val)  
                vetVal.append(val)

        else:


            driver.find_element_by_xpath(f'//a[@data-page="{i}"]').click()

            #numeros = driver.get_attribute('outerHTML')

            #valor = BeautifulSoup(test, 'html.parser')

            print()

            sleep(1)
            clas1 = driver.find_element_by_xpath('//*')

            valueCalss1 = clas1.get_attribute('outerHTML')

            value1 = BeautifulSoup(valueCalss1, 'html.parser')

            links1 = value1.find_all(class_='review_count')

            for link1 in links1:
                #for value in link:
                #value = link.find('a')
                val1 = link1.get('href')

                print(val1)   
                vetVal.append(val1)
        print(f'+++++++++++++ pagina {i} ++++++++++++')
    
    driver.close()
    return vetVal


valorLinks = []
valorLinks = linksDosEstabelecimentos(html)

print()

#print(valorLinks)


def pegaOsDados(valoresLinks):
    
    vetName = []
    vetNum = []
    vetValor = []
    
    for i in range(len(valoresLinks)):
    
        wiki = f"https://www.tripadvisor.com.br{valoresLinks[i]}"

        try:
            page1 = urllib.request.urlopen(wiki)
        except:
            pass

        # Pega o nome
        try:
            soup = BeautifulSoup(page1, 'html5lib')
            h1 = soup.find('h1')
            for name in h1:
                print(name)

                vetName.append(name)
        except:
            print('Não tem')
            vetName.append('-')

        # Pega o numero
        try:
            numero = soup.find('span', class_='_15QfMZ2L')
            for number in numero:
                for num in number:
                    print(num)

                    vetNum.append(num)
        except:
            print('Não tem')
            vetName.append('-')
            
        # Pega o $
        try:
            valor = soup.find('a', class_='_2mn01bsa')
            for value in valor:
                print(value)

                vetValor.append(value)
        except:
            print('Não tem')
            vetName.append('-')
        
        print(f'+++++++++++++++++ PEGOU O {i+1} NUMERO +++++++++++++++++')
    
    return vetName, vetNum, vetValor

vetorNome = []
vetorNumero = []
vetorValor = []

vetorNome, vetorNumero, vetorValor = pegaOsDados(valorLinks)

#print(f'Tamanho do vetor Numero -> {vetorNumero}')

for j in range(len(vetorNumero)):
    f.writerow([vetorNome[j],vetorNumero[j],vetorValor[j]])

print('FINALIZOU')
```
